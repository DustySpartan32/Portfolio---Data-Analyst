#Visualisation 1: Most Popular Game Genre

```python
top_genres = df['Genres'].value_counts().loc[lambda x: x.index != 'Unknown'].nlargest(10)
top_genres.plot(kind='pie', figsize=(8, 8), autopct='%1.1f%%')
plt.ylabel('')
plt.show()
```
